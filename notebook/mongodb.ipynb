```python
import pandas as pd
import pymongo

df = pd.read_csv("vehicle_data.csv")
df.head()
```

```python
# convert dataframe into dict format before pushing into mangodb
data = df.to_dict(orient='records')

# data
```

```python
# add parent path 
import sys
import os

sys.path.append(os.path.abspath(".."))

from credentials import MONGODB_URI

DB_NAME = "Proj1"
COLLECTION_NAME = "Proj1-Data"
CONNECTION_URL = MONGODB_URI

# above, either remove your credentials or delete the mongoDB resource bofore pushing it to github.
```

```python
# mongodb heirarchy
# project
client = pymongo.MongoClient(CONNECTION_URL)

# database 
data_base = client[DB_NAME]

# collection
collection = data_base[COLLECTION_NAME]
```

```python
# uploading data into mongoDB

# rec = collection.delete_many({}) # to avoid inserting data again (duplicates) if this cell run n times
rec = collection.insert_many(data)
```

```python
# checked for duplicates entries (if any)

pipeline = [
    {
        "$group": {
            "_id": "$id",
            "count": {"$sum": 1}
        }
    },
    {
        "$match": {
            "count": {"$gt": 1}
        }
    }
]

duplicates = list(collection.aggregate(pipeline))

print("Total duplicate IDs:", len(duplicates))
```
